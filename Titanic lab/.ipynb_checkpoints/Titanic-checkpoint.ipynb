## Step 3: Feature extraction
1.  There are two columns that pertain to how many family members are on the boat for a given person. Create a new column called `FamilyCount` which will be the sum of those two columns.
2. Reverends have a special title in their name. Create a column called `IsReverend`: 1 if they're a preacher, 0 if they're not.
3. In order to feed our training data into a classification algorithm, we need to convert our categories into 1's and 0's using `pd.get_dummies`
  - Familiarize yourself with the [`pd.get_dummies` documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html)
  - Create 3 columns: `Embarked_C`, `Embarked_Q` and `Embarked_S`. These columns will have 1's and 0's that correspond to the `C`, `Q` and `S` values in the `Embarked` column
  - Do the same thing for `Sex`
  - BONUS: Extract the title from everyone's name and create dummy columns

#### There are two columns that pertain to how many family members are on the boat for a given person. Create a new column called `FamilyCount` which will be the sum of those two columns.

```python
train_df['FamilyCount'] = train_df['SibSp'] + train_df['Parch']

train_df.head()
```

#### Reverends have a special title in their name. Create a column called IsReverend: 1 if they're a preacher, 0 if they're not.

```python
train_df['IsReverend'] = train_df['Name'].str.contains('Rev').map(lambda x : int (x))
train_df.head()
```

```python
train_df['IsReverend'].value_counts()
```

#### In order to feed our training data into a classification algorithm, we need to convert our categories into 1's and 0's using pd.get_dummies

```python
pd.get_dummies(train_df, columns = ['Embarked'])
```

```python
pd.get_dummies(train_df, columns = ['Sex'])
```

#### What was the survival rate overall?

```python
Survival_rate = train_df['Survived'].value_counts(normalize = True, ascending = True).round(2)

print('Survival (0 = No; 1 = Yes)')
Survival_rate
```

```python
train_df.groupby(['Survived'], as_index = False).agg({'Survived':'count'}).apply(lambda x:100 * x / float(x.sum())).round(2)
```

#### Which gender fared the worst? What was their survival rate?

```python
train_df[['Survived', 'Sex']].groupby('Sex', as_index = False).mean()
```

```python
train_df.groupby(['Sex','Survived'], as_index = False).agg({'Survived':'count'})
```

```python
# percentage of 
gender_rate = train_df.groupby(['Sex','Survived']).agg({'Survived':'count'}).apply(lambda x:100 * x / float(x.sum())).round(2)
gender_rate
```

#### What was the survival rate for each Pclass?

```python
train_df.groupby(['Pclass', 'Survived']).agg({'Survived':'count'}).apply(lambda x:100 * x / float(x.sum())).round(2)
```

#### Did any reverends survive? How many?

```python
train_df.query('IsReverend == 1')
```

```python
train_df.loc[(train_df['Survived'] == 1) & (train_df['IsReverend'] == 1)].count()
```

```python
train_df.filter(['Survived','IsReverend']).groupby(['IsReverend'], as_index = False).count()
```

```python
#Reverend have 6 people, there were no reverends who survived
```

#### What is the survival rate for cabins marked ¯\(ツ)/¯

```python
train_df.filter(['Survived','Cabin']).get(train_df["Cabin"] == '¯(ツ)/¯').groupby(['Survived'], as_index = False).count()
```

#### What is the survival rate for people whose Age is empty?

```python
train_df.filter(['Survived','Age']).isnull().value_counts()
```

#### What is the survival rate for each port of embarkation?

```python
embarkation_survival = train_df.get(['Survived','Embarked']).groupby(['Survived','Embarked'], as_index = False).value_counts()

embarkation_survival = pd.DataFrame(embarkation_survival) 

embarkation_survival
```

#### What is the survival rate for children (under 12) in each Pclass?

```python
train_df.get(['Survived','Age','Pclass']).where(train_df["Age"] < 12).groupby(['Survived','Age'], as_index = False).count()
```

```python
train_df.get(['Survived','Pclass']).where(train_df["Age"] < 12).groupby(['Survived'], as_index = False).count()
```

#### Did the captain of the ship survive? Is he on the list?

```python
train_df.loc[train_df['Name'].str.contains('Capt')]
```

#### Of all the people that died, who had the most expensive ticket? How much did it cost?

```python
train_df.get(['Pclass', 'Name', 'Fare','Survived']).where(train_df['Survived'] == 0).nlargest(5, 'Fare')
```

#### Does having family on the boat help or hurt your chances of survival?

```python
familymb_survival = train_df.get(['Survived','FamilyCount']).groupby('Survived', as_index = False).value_counts()

familymb_survival = pd.DataFrame(familymb_survival)
familymb_survival
```

```python
#Histogram of FamilyCount Who Survived
train_df.query('Survived == 1')[['FamilyCount']].plot(kind = 'hist');
```

```python
#Histogram of FamilyCount Who not Survived
train_df.query('Survived == 0')[['FamilyCount']].plot(kind = 'hist');
```

## Step 5: Plotting
Using Matplotlib and Seaborn, create several charts showing the survival rates of different groups of people. It's fine if a handful of charts are basic (Gender, Age, etc), but what we're really looking for is something beneath the surface.

```python
## Correlation between Survived and others variance
# Establish size of figure.
plt.figure(figsize = (8,8))

# Get correlation of variables.
corr = train_df.corr()

# Set up mask to hide upper upper triangle.
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

# Plot our correlation heatmap, while masking the upper triangle to be white.
sns.heatmap(data = corr, mask = mask, square = True, cmap = 'mako', annot = True);
```

```python
train_df.groupby(['Sex','Survived'], as_index = False).agg({'Survived':'count'})
```

```python
gender = train_df.groupby(['Sex','Survived'], as_index = False).agg({'Survived':'count'})
gender.insert(0, 'Survivel_Sex', ['Female not Survived', 'Female Survived', 'Male not Survived', 'Male Survived'], True)
gender.drop(columns = ['Sex'])
gender.plot(x = 'Survivel_Sex', y = 'Survived', kind = 'bar', color = '#2a9d8f',)
plt.title('Gender Survival');
```

```python
#Histogram of Age Who Survived
train_df.query('Survived == 1')[['Age']].plot(kind = 'hist', color = '#e9c46a');
plt.title('Histogram of Age Who Survived')

#Histogram of Age Who Did Not Survived
train_df.query('Survived == 0')[['Age']].plot(kind = 'hist', color = '#e9c46a');
plt.title('Histogram of Age Who Did Not Survived')
```
